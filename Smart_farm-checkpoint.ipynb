# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dept = [1, 5, 10, 50, 100, 500,800, 1000]
min_samples =  [5, 10, 100, 500]


param_grid={'min_samples_split':min_samples , 'max_depth':dept}
clf = DecisionTreeClassifier()
model = GridSearchCV(clf,param_grid,scoring='accuracy',n_jobs=-1,cv=3)
model.fit(x_train,y_train)
print("optimal min_samples_split",model.best_estimator_.min_samples_split)
print("optimal max_depth",model.best_estimator_.max_depth)
```

```python
from sklearn.metrics import accuracy_score
dt = DecisionTreeClassifier(criterion='entropy',splitter='best',max_depth =50,min_samples_split =5)
dt.fit(x_train,y_train)

import pickle
filename = r'C:\Users\dsp96\Desktop\SMART_FARM\FRONT END\dt_smart.pkl'
pickle.dump(dt, open(filename, 'wb'))

pred_test1 =dt.predict(x_test)
test_accuracy = accuracy_score(y_test, pred_test1)
pred_train1 = dt.predict(x_train)
train_accuracy =accuracy_score(y_train,pred_train1)



print("Accuracy for Test data is " +str(accuracy_score(y_test,pred_test1)))
print("Accuracy for Train data is " +str(accuracy_score(y_train,pred_train1)))

print("---------------------------")
class_names =['Clay-Drip Irrigation', 'Saline-Drip Irrigation',
       'Clay-Sprinkler Irrigation', 'Black lava soil-Tube Well',
       'Chalky (Calcareous)-Sprinkler Irrigation',
       'Sandy-Drip Irrigation', 'Chalky (Calcareous)-Canal Irrigation',
       'Nitrogenous-Drip Irrigation', 'Saline-Tube Well',
       'Nitrogenous-Tube Well', 'Nitrogenous-Canal Irrigation',
       'Nitrogenous-Sprinkler Irrigation', 'Alkaline-Drip Irrigation',
       'Black lava soil-Sprinkler Irrigation',
       'Sandy-Sprinkler Irrigation', 'Loamy-Tube Well',
       'Black lava soil-Canal Irrigation', 'Clay-Tube Well',
       'Chalky (Calcareous)-Tube Well', 'Saline-Canal Irrigation',
       'Loamy-Drip Irrigation', 'Sandy-Tube Well',
       'Loamy-Canal Irrigation', 'Alkaline-Canal Irrigation',
       'Alkaline-Tube Well', 'Black lava soil-Drip Irrigation',
       'Alkaline-Sprinkler Irrigation', 'Sandy-Canal Irrigation',
       'Loamy-Sprinkler Irrigation', 'Saline-Sprinkler Irrigation',
       'Clay-Canal Irrigation', 'Chalky (Calcareous)-Drip Irrigation']
df_heatmap = pd.DataFrame(confusion_matrix(y_test, pred_test1.round()), index=class_names, columns=class_names )
fig = plt.figure( )
heatmap = sns.heatmap(df_heatmap, annot=True, fmt="d")
```
